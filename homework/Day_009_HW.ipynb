# 檢視與處理 Outliers
### 為何會有 outliers, 常見的 outlier 原因
* 未知值，隨意填補 (約定俗成的代入)，如年齡常見 0,999
* 可能的錯誤紀錄/手誤/系統性錯誤，如某本書在某筆訂單的銷售量 = 1000 本

# [作業重點]

- 繪製目標值累積密度函數(ECDF)的圖形, 和常態分布的累積密度函數對比, 以確認是否有離群值的情形 (In[6], Out[6], In[7], Out[7])

```python
# Import 需要的套件
import os
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt

%matplotlib inline

# 設定 data_path
dir_data = '.\data'
```

```python
f_app = os.path.join(dir_data, 'application_train.csv')
print('Path of read in data: %s' % (f_app))
app_train = pd.read_csv(f_app)
app_train.head()
```

## 請參考 HomeCredit_columns_description.csv 的欄位說明，觀察並列出三個你覺得可能有 outlier 的欄位並解釋可能的原因

```python
# 先篩選數值型的欄位
#print(app_train.dtypes)
dtype_select = [np.dtype('float64'), np.dtype('int64')]
numeric_columns = list(app_train.columns[list(app_train.dtypes.isin(dtype_select))]) #dtype_select = numpy float,numpy int

# 再把只有 2 值 (通常是 0,1) 的欄位去掉
numeric_columns = list(app_train[numeric_columns].columns[list(app_train[numeric_columns].apply(lambda x:len(x.unique())!=2 ))])

print("Numbers of remain columns ",len(numeric_columns))

for col in numeric_columns:
    app_train[col].plot.hist(title = col)
    plt.show()
#可疑的有 DAYS_EMOLOYED 、 AMT_INCOME_TOTAL 有小數，都要整數 、 REGION_POPULATION_RELATIVE 本欄位為人口數，數字為整數非小數
#OBS_60_CNT_SOCIAL_CIRCLE、OBS_30_CNT_SOCIAL_CIRCLE、DEF_30_CNT_SOCIAL_CIRCLE、YEARS_BEGINEXPLUATATION_MEDI、YEARS_BEGINEXPLUATATION_MODE
#資料量夠多會變常態分佈
```

### Hints: Emprical Cumulative Density Plot, [ECDF](https://zh.wikipedia.org/wiki/%E7%BB%8F%E9%AA%8C%E5%88%86%E5%B8%83%E5%87%BD%E6%95%B0), [ECDF with Python](https://stackoverflow.com/questions/14006520/ecdf-in-python-without-step-function)

```python
# 最大值離平均與中位數很遠
print(app_train['AMT_INCOME_TOTAL'].describe())
```

```python
import statsmodels.api as sm
# 繪製 Empirical Cumulative Density Plot (ECDF)

cdf = sm.distributions.ECDF(app_train['AMT_INCOME_TOTAL'])

plt.plot(cdf.x,cdf.y/cdf.y.max())
plt.xlabel('Value')
plt.ylabel('ECDF')
#plt.xlim([cdf.x.min(), cdf.x.max() * 1.05]) # 限制顯示圖片的範圍
plt.xlim([app_train['AMT_INCOME_TOTAL'].min(), app_train['AMT_INCOME_TOTAL'].max()]) # 限制顯示圖片的範圍
plt.ylim([-0.05,1.05]) # 限制顯示圖片的範圍

plt.show()

plt.plot(np.log(cdf.x),cdf.y)
plt.xlabel('Value (log-scale)')
plt.ylabel('ECDF')

plt.ylim([-0.05,1.05]) # 限制顯示圖片的範圍

plt.show()
```

## 補充：Normal dist 的 ECDF

```python
# 最大值落在分布之外
print(app_train['REGION_POPULATION_RELATIVE'].describe())
```

```python
import statsmodels.api as sm
# 繪製 Empirical Cumulative Density Plot (ECDF)

cdf = sm.distributions.ECDF(app_train['REGION_POPULATION_RELATIVE'])


plt.plot(cdf.x,cdf.y/cdf.y.max())
plt.xlabel('Value')
plt.ylabel('ECDF')
plt.ylim([-0.05,1.05]) # 限制顯示圖片的範圍
plt.show()

app_train['REGION_POPULATION_RELATIVE'].hist()
plt.show()

app_train['REGION_POPULATION_RELATIVE'].value_counts()

# 就以這個欄位來說，雖然有資料掉在分布以外，也不算異常，僅代表這間公司在稍微熱鬧的地區有的據點較少，
# 導致 region population relative 在少的部分較為密集，但在大的部分較為疏漏
```

```python
# 最大值落在分布之外
print(app_train['OBS_60_CNT_SOCIAL_CIRCLE'].describe())
```

```python
import statsmodels.api as sm

# 繪製 Empirical Cumulative Density Plot (ECDF)

cdf = sm.distributions.ECDF(app_train['OBS_60_CNT_SOCIAL_CIRCLE'])



plt.plot(cdf.x,cdf.y/cdf.y.max())
plt.xlabel('Value')
plt.ylabel('ECDF')
#plt.xlim([cdf.x.min() * 0.95, cdf.x.max() * 1.05])
plt.xlim([app_train['OBS_60_CNT_SOCIAL_CIRCLE'].min(), app_train['OBS_60_CNT_SOCIAL_CIRCLE'].max()]) # 限制顯示圖片的範圍
plt.ylim([-0.05,1.05]) # 限制顯示圖片的範圍
plt.show()

app_train['OBS_60_CNT_SOCIAL_CIRCLE'].hist()
plt.show()
print(app_train['OBS_60_CNT_SOCIAL_CIRCLE'].value_counts().sort_index(ascending = False))
```
